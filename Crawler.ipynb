```python
import numpy as np
import pandas as pd
import cassiopeia as cass
import datetime
import random
```

```python
# Create a dataframe with single row to keep track of stats and then append to the stat_import dataframe which will then append to the main dataframe or csv

new_df = pd.DataFrame()
new_df['First_Bloods'] = [0]
new_df['Double_Kills'] = [0]
new_df['Triple_Kills'] = [0] 
new_df['Quadra_Kills'] = [0] 
new_df['Penta_Kills'] = [0] 
new_df['KDA']= [0] 
new_df['Kills']= [0]
new_df['Deaths']= [0] 
new_df['Assists']= [0] 
new_df['CS']= [0] 
new_df['Damage_Dealt'] = [0] 
new_df['Damage_Taken'] = [0] 
new_df['Gold_Earned']= [0] 
new_df['Turret_Damage']= [0] 
new_df['Turret_Kills']= [0] 
new_df['Turrt_Takedowns']= [0]
new_df['VisionW_Bought']= [0] 
new_df['VisionW_Placed']= [0] 
new_df['Wards_Placed']= [0] 
new_df['Wards_Killed']= [0] 
new_df['CC_Score']= [0] 
```

```python
# This function will get all the stats from a game and then append it to the new_df, calulations will be performed after data collectionis completed

def get_stats(player):
    first_blood = player.stats.first_blood_kill
    double_kill = player.stats.double_kills
    triple_kill = player.stats.triple_kills
    quadra_kills = player.stats.quadra_kills
    penta_kills = player.stats.penta_kills
    kda = player.stats.kda
    kills = player.stats.kills
    deaths = player.stats.deaths
    assists = player.stats.assists
    CS = player.stats.total_minions_killed + player.stats.neutral_minions_killed
    damage_dealt = player.stats.total_damage_dealt_to_champions
    damage_taken = player.stats.total_damage_taken
    gold_earned = player.stats.gold_earned
    turret_damage = player.stats.damage_dealt_to_turrets
    turret_kills = player.stats.turret_kills
    turret_takedowns = player.stats.turret_takedowns
    vision_wards_bought = player.stats.vision_wards_bought 
    vision_wards_placed = player.stats.vision_wards_placed 
    wards_placed = player.stats.wards_placed 
    wards_killed = player.stats.wards_killed 
    cc_score = player.stats.time_CCing_others



    new_df['First_Bloods'] += first_blood
    new_df['Double_Kills'] += double_kill
    new_df['Triple_Kills'] += triple_kill 
    new_df['Quadra_Kills'] += quadra_kills 
    new_df['Penta_Kills'] += penta_kills 
    new_df['KDA'] += kda 
    new_df['Kills'] += kills
    new_df['Deaths'] += deaths 
    new_df['Assists'] += assists 
    new_df['CS'] += CS 
    new_df['Damage_Dealt'] += damage_dealt 
    new_df['Damage_Taken'] += damage_taken 
    new_df['Gold_Earned'] += gold_earned 
    new_df['Turret_Damage'] += turret_damage 
    new_df['Turret_Kills'] += turret_kills 
    new_df['Turrt_Takedowns'] += turret_takedowns
    new_df['VisionW_Bought'] += vision_wards_bought 
    new_df['VisionW_Placed'] += vision_wards_placed 
    new_df['Wards_Placed'] += wards_placed 
    new_df['Wards_Killed'] += wards_killed 
    new_df['CC_Score'] += cc_score 
        
```

```python
# create an empty dataframe to hold stats

stat_import = pd.DataFrame()
```

```python
# this function will take in the ID of the seed acount and the amount of games we want from their match history 

def add_match_history(puuid, games):
    global df, df2, top_stats, jg_stats, mid_stats, bot_stats, sup_stats, MATCHES, print_index


    # call their match history and get the amount of games from the parameter, 8,10,12 etc. 
    match_history = cass.MatchHistory(puuid=puuid,continent="AMERICAS", region='NA',begin_index=0,end_index=games,queue=cass.Queue.ranked_solo_fives)
    
    # loop through all matches in the match_history 
    for match in match_history:

        # check if the ID of the match is already in the main csv
        try:
            if str(match.id) in MATCHES["ID"].values or str(match.id)[4:] in MATCHES["ID"].values:
                continue
        except:
            pass
        
        # if the game was less than 15 minutes we do not want to keep track of it, (early surrender because someone was afk most likely)
        if match.duration < datetime.timedelta(hours=0, minutes=15, seconds=0):
            continue
        
        # create ID column in stat_import
        stat_import['ID']=[0]      

        # add the id of this match 
        if type(match.id) == str:
            stat_import['ID'] = match.id[4:]
        else:
            stat_import['ID'] = str(match.id)

        # create a list to hold players of each team
        player_list = [] 
        player_list2 = []   


        # random number generator to decide which team to keep track of so we can get games from both map sides
        random_number = random.randint(1, 2)

        # if side is 1 then the blue team will be "our" team
        if(random_number == 1):
            stat_import['Win'] = [match.blue_team.win]
            player_list.append(match.participants[0])
            player_list.append(match.participants[1])
            player_list.append(match.participants[2])
            player_list.append(match.participants[3])
            player_list.append(match.participants[4])
            stat_import['Side'] = ['Blue']
            player_list2.append(match.participants[5])
            player_list2.append(match.participants[6])
            player_list2.append(match.participants[7])
            player_list2.append(match.participants[8])
            player_list2.append(match.participants[9])

        # if side is 2 then "red" team will be "our" team
        elif(random_number == 2):
            stat_import['Win'] = [match.red_team.win]
            player_list.append(match.participants[5])
            player_list.append(match.participants[6])
            player_list.append(match.participants[7])
            player_list.append(match.participants[8])
            player_list.append(match.participants[9])
            stat_import['Side'] = ['Red']
            player_list2.append(match.participants[0])
            player_list2.append(match.participants[1])
            player_list2.append(match.participants[2])
            player_list2.append(match.participants[3])
            player_list2.append(match.participants[4])

        # create columns to store all the data of the players for this match
        stat_import['Rank'] = ['Diamond']
        stat_import['Top_Current_Champ'] = [0]
        stat_import['Top_games_on_role'] = [0]
        stat_import['Top_games_on_champ'] = [0]
        stat_import['Top_Win'] = [0]
        stat_import['Top_Lose'] = [0]
        stat_import['Top_Win_2'] = [0]
        stat_import['Top_Win_5'] = [0]
        stat_import['Jg_Current_Champ'] = [0]
        stat_import['Jg_games_on_role'] = [0]
        stat_import['Jg_games_on_champ'] = [0]
        stat_import['Jg_Win'] = [0]
        stat_import['Jg_Lose'] = [0]
        stat_import['Jg_Win_2'] = [0]
        stat_import['Jg_Win_5'] = [0]
        stat_import['Mid_Current_Champ'] = [0]  
        stat_import['Mid_games_on_role'] = [0]
        stat_import['Mid_games_on_champ'] = [0]
        stat_import['Mid_Win'] = [0]
        stat_import['Mid_Lose'] = [0]
        stat_import['Mid_Win_2'] = [0]
        stat_import['Mid_Win_5'] = [0]
        stat_import['Bot_Current_Champ'] = [0]
        stat_import['Bot_games_on_role'] = [0]
        stat_import['Bot_games_on_champ'] = [0]
        stat_import['Bot_Win'] = [0]
        stat_import['Bot_Lose'] = [0]
        stat_import['Bot_Win_2'] = [0]
        stat_import['Bot_Win_5'] = [0]
        stat_import['Sup_Current_Champ'] = [0]
        stat_import['Sup_games_on_role'] = [0]
        stat_import['Sup_games_on_champ'] = [0]
        stat_import['Sup_Win'] = [0]
        stat_import['Sup_Lose'] = [0]
        stat_import['Sup_Win_2'] = [0]
        stat_import['Sup_Win_5'] = [0]

        # enemy team champions
        stat_import['Top_Enemy_Champ'] = [0]
        stat_import['Jg_Enemy_Champ'] = [0]
        stat_import['Mid_Enemy_Champ'] = [0]
        stat_import['Bot_Enemy_Champ'] = [0]
        stat_import['Sup_Enemy_Champ'] = [0]

        # player list two is the team that counts as the enemy team, add their champions to the dataframe
        for player in player_list2:
            if player.individual_position.name == 'top_lane':
                stat_import['Top_Enemy_Champ'] = [player.champion.name]
            if player.individual_position.name == 'jungle':
                stat_import['Jg_Enemy_Champ'] = [player.champion.name]
            if player.individual_position.name == 'mid_lane':
                stat_import['Mid_Enemy_Champ'] = [player.champion.name]
            if player.individual_position.name == 'bot_lane':
                stat_import['Bot_Enemy_Champ'] = [player.champion.name]
            if player.individual_position.name == 'utility':
                stat_import['Sup_Enemy_Champ'] = [player.champion.name]

        # player list one is "our" team, add their champions to the data frame
        for player in player_list:
            if player.individual_position.name == 'top_lane':
                stat_import['Top_Current_Champ'] = [player.champion.name]
                puuid_one = player.summoner.puuid
            if player.individual_position.name == 'jungle':
                stat_import['Jg_Current_Champ'] = [player.champion.name]
                puuid_two = player.summoner.puuid
            if player.individual_position.name == 'mid_lane':
                stat_import['Mid_Current_Champ'] = [player.champion.name]
                puuid_three = player.summoner.puuid
            if player.individual_position.name == 'bot_lane':
                stat_import['Bot_Current_Champ'] = [player.champion.name]
                puuid_four = player.summoner.puuid
            if player.individual_position.name == 'utility':
                stat_import['Sup_Current_Champ'] = [player.champion.name]
                puuid_five = player.summoner.puuid

        # store the time this match was created
        match_creation = match.creation.int_timestamp

        # get match histories of all players on "our" team begin time is january, end time is the creation time of the current match
        top_history = cass.MatchHistory(puuid=puuid_one,continent="AMERICAS", region='NA',begin_time=1641600000,end_time=match_creation,queue=cass.Queue.ranked_solo_fives)
        jg_history = cass.MatchHistory(puuid=puuid_two,continent="AMERICAS", region='NA',begin_time=1641600000,end_time=match_creation,queue=cass.Queue.ranked_solo_fives)
        mid_history = cass.MatchHistory(puuid=puuid_three,continent="AMERICAS", region='NA',begin_time=1641600000,end_time=match_creation,queue=cass.Queue.ranked_solo_fives)
        bot_history = cass.MatchHistory(puuid=puuid_four,continent="AMERICAS", region='NA',begin_time=1641600000,end_time=match_creation,queue=cass.Queue.ranked_solo_fives)
        sup_history = cass.MatchHistory(puuid=puuid_five,continent="AMERICAS", region='NA',begin_time=1641600000,end_time=match_creation,queue=cass.Queue.ranked_solo_fives)


        # This will loop through all five match histories and add their stats to the dataframe
        # as its looping it will sum their wins
        # then it will call the get_stats function to get everything else
        # after its done looping it will divide some columns by 20 games to average out the stats

        try:
            match_count = 0
            for matches in top_history:
                
                for participants in matches.participants:
                       

                    if participants.summoner.puuid == puuid_one:    
                        if participants.individual_position.name == 'top_lane':
                            stat_import['Top_games_on_role'] += 1
                        if participants.champion.name == stat_import.iloc[0]['Top_Current_Champ']:
                            stat_import['Top_games_on_champ'] += 1

                        if participants.stats.win.real == 1:
                            stat_import['Top_Win'] += 1
                        elif participants.stats.win.real == 0:
                            stat_import['Top_Lose'] += 1

                        if match_count < 2:
                            if participants.stats.win.real == 1:
                                stat_import['Top_Win_2'] += 1
                                
                        if match_count < 5:
                            if participants.stats.win.real == 1:
                                stat_import['Top_Win_5'] += 1

                        match_count += 1

                        get_stats(participants)
                        break
        except:
            pass
        new_df.iloc[0:,5:] = new_df.iloc[0:,5:] / 20
        for column in new_df.columns:
            stat_import['Top_'+column] = new_df[column] 
            new_df[column] = [0] 
        try:
            match_count = 0
            for matches in jg_history:
                for participants in matches.participants:
                    if participants.summoner.puuid == puuid_two:
                        if participants.individual_position.name == 'jungle':
                            stat_import['Jg_games_on_role'] += 1
                        if participants.champion.name == stat_import.iloc[0]['Jg_Current_Champ']:
                            stat_import['Jg_games_on_champ'] += 1
                        if participants.stats.win.real == 1:
                            stat_import['Jg_Win'] += 1
                        elif participants.stats.win.real == 0:
                            stat_import['Jg_Lose'] += 1
                        if match_count < 2:
                            if participants.stats.win.real == 1:
                                stat_import['Jg_Win_2'] += 1
                        if match_count < 5:
                            if participants.stats.win.real == 1:
                                stat_import['Jg_Win_5'] += 1

                        match_count += 1

                        get_stats(participants)
                        break
        except:
            pass
        new_df.iloc[0:,5:] = new_df.iloc[0:,5:] / 20
        for column in new_df.columns:
                    stat_import['Jg_'+column] = new_df[column]
                    new_df[column] = [0]
        try:
            match_count = 0
            for matches in mid_history:
                for participants in matches.participants:
                    if participants.summoner.puuid == puuid_three:
                        if participants.individual_position.name == 'mid_lane':
                            stat_import['Mid_games_on_role'] += 1
                        if participants.champion.name == stat_import.iloc[0]['Mid_Current_Champ']:
                            stat_import['Mid_games_on_champ'] += 1
                        if participants.stats.win.real == 1:
                            stat_import['Mid_Win'] += 1
                        elif participants.stats.win.real == 0:
                            stat_import['Mid_Lose'] += 1
                        if match_count < 2:
                            if participants.stats.win.real == 1:
                                stat_import['Mid_Win_2'] += 1
                        if match_count < 5:
                            if participants.stats.win.real == 1:
                                stat_import['Mid_Win_5'] += 1
                        match_count += 1
                        get_stats(participants)
                        break
        except:
            pass
        new_df.iloc[0:,5:] = new_df.iloc[0:,5:] / 20
        for column in new_df.columns:
            stat_import['Mid_'+column] = new_df[column]
            new_df[column] = [0] 
        try:
            match_count = 0
            for matches in bot_history:
                for participants in matches.participants:
                    if participants.summoner.puuid == puuid_four:
                        if participants.individual_position.name == 'bot_lane':
                            stat_import['Bot_games_on_role'] += 1
                        if participants.champion.name == stat_import.iloc[0]['Bot_Current_Champ']:
                            stat_import['Bot_games_on_champ'] += 1
                        if participants.stats.win.real == 1:
                            stat_import['Bot_Win'] += 1
                        elif participants.stats.win.real == 0:
                            stat_import['Bot_Lose'] += 1
                        if match_count < 2:
                            if participants.stats.win.real == 1:
                                stat_import['Bot_Win_2'] += 1
                        if match_count < 5:
                            if participants.stats.win.real == 1:
                                stat_import['Bot_Win_5'] += 1
                        match_count += 1
                        get_stats(participants)
                        break
        except:
            pass
        new_df.iloc[0:,5:] = new_df.iloc[0:,5:] / 20
        for column in new_df.columns:
            stat_import['Bot_'+column] = new_df[column] 
            new_df[column] = [0] 
        try:
            match_count = 0
            for matches in sup_history:
                for participants in matches.participants:
                    if participants.summoner.puuid == puuid_five:
                        if participants.individual_position.name == 'utility':
                            stat_import['Sup_games_on_role'] += 1
                        if participants.champion.name == stat_import.iloc[0]['Sup_Current_Champ']:
                            stat_import['Sup_games_on_champ'] += 1
                        if participants.stats.win.real == 1:
                            stat_import['Sup_Win'] += 1
                        elif participants.stats.win.real == 0:
                            stat_import['Sup_Lose'] += 1
                        if match_count < 2:
                            if participants.stats.win.real == 1:
                                stat_import['Sup_Win_2'] += 1
                        if match_count < 5:
                            if participants.stats.win.real == 1:
                                stat_import['Sup_Win_5'] += 1
                        match_count += 1
                        get_stats(participants)
                        break
        except:
            pass
        new_df.iloc[0:,5:] = new_df.iloc[0:,5:] / 20
        for column in new_df.columns:
            stat_import['Sup_'+column] = new_df[column] 
            new_df[column] = [0] 

        # append this match to the df used by the loop to skip duplicates
        MATCHES = MATCHES.append(stat_import, ignore_index = True)

        
        # append to a csv so if the program errors out, all the data is saved and can be appended later
        if print_index == 0:
            print("OUTPUTTING TO CSV")
            stat_import.to_csv("matches.csv",index=False,mode='w',header=True)
            print_index+=1
        else:
            stat_import.to_csv("matches.csv",index=False,mode='a',header=False)

    return find_next_player(match_history)
    
```

```python
# Function that selects a random player from the last player's previous games
def find_next_player(history):
    game = 0
    New_Player_Attempts = 0
    while True:
        New_Player_Attempts += 1
        # the match to select a new player from
        match = history[game]
        # selecting random number to get the random player from match
        randNum = np.random.randint(10)
        player = match.participants[randNum]
        # If player has already been checked
        if player.summoner.name in already_checked:
            # only 10 players in a match, if all players been checked go to next game
            if New_Player_Attempts == 10:
                game += 1
                New_Player_Attempts = 0
                continue
            else:
                continue
        break
    print(player.summoner.name)
    # keep track of the account that has been checked
    already_checked.append(player.summoner.name)
    # Returns the puuid of the new player
    return player.summoner.puuid
```
